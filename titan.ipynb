```python
import pandas as pd
titan = pd.read_csv('/content/drive/MyDrive/titanic/titanic_train.csv')
titan.head()
```

```python
titan= titan.drop(['PassengerId', 'Name' ,'Cabin' ] ,axis = 1)
titan
```

```python
titan.isnull().sum()
```

```python
titan['Age'] = titan['Age'].interpolate(method = 'linear')
```

```python
titan['Embarked'] = titan['Embarked'].fillna(titan['Embarked'].mode()[0])
```

```python
titan.isnull().sum()
```

```python
titan['Embarked'].value_counts()
```

```python
def func(s):
  if s == 'S':
    return 0
  elif s == "C":
    return 1
  else:
    return 2
titan['Embarked'] = titan['Embarked'].apply(func)
```

```python
titan.head(10)
```

```python
def func1(s):
  if s == 'male':
    return 0
  else:
    return 1
titan['Sex'] = titan['Sex'].apply(func1)
```

```python
titan.head()
```

```python
titan = titan.drop(['Ticket'] ,axis = 1)
titan.head()
```

```python
x = titan.drop(['Survived'] ,axis = 1)
y = titan['Survived']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
pred
```

```python
x_test
```

```python
from sklearn.svm import SVC
model1 = SVC()
model1.fit(x_train, y_train)
```

```python
pred1 = model1.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
```

```python
pred2 = model2.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred2)
```
